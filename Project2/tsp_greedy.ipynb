# Greedy Algorithm Pseudocode:

# Helper function 1
define function load_data:
    create two empty lists for storing names and distances

    open the names file for reading
    for each line in the file:
        remove the newline character
        add the cleaned-up city name to the list of cities

    open the distance file for reading
    for each line in the file:
        remove the newline character
        split the line into individual numbers (as strings)
        convert each string to a number
        add the list of numbers to the distances list
        
        return both lists (cities and distances)

# Helper function 2
 define function nearest_unvisited:
    set closest_city to nothing yet
    set shortest_trip to a large placeholder number

    for each city still unvisited:
        look up the distance from the current city to this unvisited city
        if this distance < current shortest_trip:
            update shortest_trip with new shorter distance
            store this city as closest_city

    return closest city found and shortest_trip distance

# Main function
define function:
    call the load_data function to read the city names and distance values

    create variables to track the best route and the shortest total distance

    for loop for each starting city:
        route = list starting with the current starting city
        distance_traveled = 0
        unvisited_cities = list of all city indexes except the starting city

        inner loop while there are still unvisited cities:
            current_city = last city added to the route
            call nearest_unvisited to find the closest unvisited city and its distance
            add that city to the route
            add the distance to the total distance
            remove that city from the unvisited list

        after visiting all cities:
            add distance from the last city back to the starting city

        if this route’s total distance is shorter than the current best:
            update best_route and shortest_distance with this new route

    convert best_route from index numbers to city names

    return the final route and the total distance

````python
# Constant
LONG = 99999999
````

````python
# Helper Function 1
def read_files(name_file, distance_file):
    
    with open(name_file, "r") as read_file:
        cities = [city.strip() for city in read_file]
    
    with open(distance_file, "r") as read_file:
        dist_values = [[float(dist) for dist in distances.split()] for distances in read_file]
          
    return cities, dist_values
````

````python
# Helper Function 2
def closest_unvisited(current_city, unvisited_cities, distances):
    closest_city = None
    shortest_trip = LONG
            
    for next_stop in unvisited_cities:
        if distances[current_city][next_stop] < shortest_trip:
            shortest_trip = distances[current_city][next_stop]
            closest_city = next_stop
    
    return closest_city, shortest_trip
````

````python
# Main Function
def tsp_greedy(name_file, distance_file):
    
    # calling helper function 1
    cities, distances = read_files(name_file, distance_file)
    
    shortest_distance = LONG
    best_route = []
    
    for starting_city in range(len(cities)):
        route = [starting_city]
        distance_traveled = 0.0
        unvisited_cities = list(range(len(cities)))
        unvisited_cities.remove(starting_city)
        
        # inner loop
        while unvisited_cities:
            current_city = route[-1] # points to last item in the list
            
            # calling helper function 2
            closest_city, shortest_trip = closest_unvisited(current_city, unvisited_cities, distances)
            
            route.append(closest_city)
            distance_traveled += shortest_trip
            unvisited_cities.remove(closest_city)
            
        # returning to trip starting point
        distance_traveled += distances[route[-1]][starting_city]
        
        # updating if there is a shortest distance this time
        if distance_traveled < shortest_distance:
            shortest_distance = distance_traveled
            best_route = route
    
    final_route = [cities[i] for i in best_route]
    
    return final_route, shortest_distance
````
